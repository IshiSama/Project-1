```python
def find_part(var1: str, var2: str) -> str:
    """
    Returns:
    - var1 if var1 is a substring of var2
    - var2 if var2 is a substring of var1
    - "neither variable is a part of other" otherwise
    """
    if var1 in var2:
        return var1
    elif var2 in var1:
        return var2
    else:
        return "neither variable is a part of other"


def calculate_score(transaction_amount: float, transaction_count: int) -> int:
    """
    Returns score based on:
    - 0   -> amount < 1000 AND count < 5
    - 50  -> 1000 <= amount < 10000 AND 5 <= count < 10
    - 100 -> otherwise
    """
    if transaction_amount < 1000 and transaction_count < 5:
        return 0
    elif 1000 <= transaction_amount < 10000 and 5 <= transaction_count < 10:
        return 50
    else:
        return 100


# ---- Example Usage ----
if __name__ == "__main__":
    # Part A
    var1 = "sat@google.com"
    var2 = "google.com"
    result_a = find_part(var1, var2)
    print("Result A:", result_a)

    # Part B
    transaction_amount = 5000
    transaction_count = 7
    result_b = calculate_score(transaction_amount, transaction_count)
    print("Result B:", result_b)
```
